2. processed
3. lookup

#### Please update the following 
- application-id
- service-credential
- directory-id

```
configs = {"fs.azure.account.auth.type": "OAuth",
           "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
           "fs.azure.account.oauth2.client.id": "2404b623-7409-453a-8a4c-5859b6e5510f",
           "fs.azure.account.oauth2.client.secret": "eWk7Q~2Hp0pEVP-H8W2rRWDggf4~XE9-hcxLD",
           "fs.azure.account.oauth2.client.endpoint": "https://login.microsoftonline.com/c8a207e7-6237-4549-ad92-61fcb6526d81/oauth2/token"}
```

### Mount the raw container
#### Update the storage account name before executing

```
dbutils.fs.mount(
  source = "abfss://raw@covidreportdlakes.dfs.core.windows.net/",
  mount_point = "/mnt/covidreportdlakes/raw",
  extra_configs = configs)
```
